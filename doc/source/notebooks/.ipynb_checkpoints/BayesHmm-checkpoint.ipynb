```julia
using Turing, Distributions
using ForwardDiff: Dual
using PyPlot, PyCall
```

```julia
K = 3

obs = [ 1.0, 1.0, 1.0, 1.0, 5.0, 5.0, 5.0, -4.0, -4.0, -4.0 ];
N   = length(obs);
```

```julia
@model BayesHmm(y) = begin
    s = tzeros(Int64, N)
    m = tzeros(Dual, K)
    T = Array{Array}(K)
    for i = 1:K
        T[i] ~ Dirichlet(ones(K))
    end
    m[1] ~ Normal(1, 0.1)
    m[2] ~ Normal(5, 0.1)
    m[3] ~ Normal(-4, 0.1)
    s[1] ~ Categorical(ones(Float64, K)/K)
    for i = 2:N
        # s[i] ~ Categorical(vec(T[s[i-1],:]))
        s[i] ~ Categorical(vec(T[s[i-1]]))
        y[i] ~ Normal(m[s[i]], 0.1)
    end
    res = m[s[:]]
    return(res, s, m)
end
```

```julia
g = Gibbs(300, HMC(2, 0.2, 5, :m, :T), PG(10, 2, :s))
c = @sample(BayesHmm(obs), g);
```

```julia
s = MambaChains(c);
```

```julia
describe(s)
```

```julia
m = c[:m][300];
s = c[:s][300];
PyPlot.plot(obs, linestyle="None", marker="+", color = "r")
PyPlot.plot(m[s], linestyle="-", marker="*", color = "b")
```
