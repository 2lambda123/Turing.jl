```julia
using Turing, Distributions
using Reactive, Interact, Gadfly, Compose
```

```julia
M = 3 # number of clusters
N = 100 # number of data points 
srand(124)


mu_true = [2., 2.]
Sig_true = eye(2)*20
Sig_true[1,2] = Sig_true[2,1] = 0.25

mus = rand(MultivariateNormal(mu_true, Sig_true), M)

@model gmm() = begin

    xs=Array{Vector}(N)
    cs = tzeros(Int, N)
    for i = 1:N
        cs[i] ~ Categorical(M)
        xs[i] ~ MvNormal(mus[:,cs[i]], [1/2 0; 0 1/2])
    end
    return(mus, cs, xs)
end

chain = sample(gmm, PG(10, 2))
xs = chain.value[1].value[:xs]
# Plot data points
d_layer = layer(x=map(e -> e[1], xs), y=map(e -> e[2], xs), Geom.point)

plot(d_layer, Coord.Cartesian(xmin=-7, xmax=7, ymin=-7, ymax=7))
```

```julia
N = length(xs)
M = 3

@model gmm begin
    #mus = TArray(Vector, M)
    #for i = 1:M
    #    @assume mus[i] ~ MvNormal([0 0], [20 0; 0 20])
    #end
    
    cs = tzeros(Int, N)
    for i = 1:N
        @assume cs[i] ~ Categorical(M)
        @observe xs[i] ~ MvNormal(mus[:,cs[i]], [1/2 0; 0 1/2])
    end
    @predict mus cs
end
```

```julia
chain = sample(gmm, PG(50, 100));

mus = mean(chain[:mus]);
cs = mean(chain[:cs]);
```
